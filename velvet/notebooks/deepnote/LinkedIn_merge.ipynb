```
import pandas as pd
```

```
df = pd.read_excel('./data/LinkedIn_postings/2021-01-22-job-offer-data.xlsx', 
                    index_col=0)
```

```
df.head()
```

```
# Convert to all strings and get everything but descriptions
tmp = df.astype(str)
merged = tmp.iloc[:, 0:7]
merged.head(5)
```

```
# Mush descriptions back together
# Should consider loss of structural info
merged['description'] = ''
descriptions = tmp.iloc[:, 7:]
for index, row in descriptions.iterrows():
    description = ""
    
    for x in row:
        if x != 'nan':
            description += (x + '\n')
    merged.at[index, 'description'] = description
```

```
merged['description']
```

```
merged.columns
```

```
columns = ['company','title','location', 'description']
merged = merged[['Company name', 'Job title', 'Job location', 'description']]
merged.columns = columns
```

```
# Yay, the final df - we can append this to the velvet data now
merged
```
